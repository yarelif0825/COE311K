```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(num_samples):
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    inside = (x**2 + y**2) <= 1
    return 4 * np.sum(inside) / num_samples

num_runs = 100
num_samples = 1000
pi_actual = np.pi

absolute_errors = []
relative_errors = []

for _ in range(num_runs):
    pi_est = monte_carlo_pi(num_samples)
    abs_error = abs(pi_actual - pi_est)
    rel_error = abs((pi_actual - pi_est) / pi_actual)
    absolute_errors.append(abs_error)
    relative_errors.append(rel_error)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, num_runs + 1), absolute_errors, marker='o')
plt.title('Absolute Error per Run')
plt.xlabel('Run')
plt.ylabel('Absolute Error')

plt.subplot(1, 2, 2)
plt.plot(range(1, num_runs + 1), relative_errors, marker='o', color='orange')
plt.title('Relative Error per Run')
plt.xlabel('Run')
plt.ylabel('Relative Error')

plt.tight_layout()
plt.show()
```
