```python
import numpy as np

# Leibniz series for pi

approx_pi = 0
num_terms = 5000

print("Term\tRelative Error")
for n in range(num_terms + 1):
    approx_pi += ((-1) ** n) / (2 * n + 1)
    if n % 100 == 0 and n > 0:
        current_pi = 4 * approx_pi
        rel_error = abs((current_pi - np.pi) / np.pi)
        print(f"{n}\t{rel_error:.8f}")
        
# As each term is added to the summation, the relative error decreases and the approximation gets closer to the true value of pi. 
```
