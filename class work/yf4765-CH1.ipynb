```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0, 2 * np.pi, 50) 
rand_data = np.random.rand(1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, np.sin(x), 'r-o', label='sin(x)')
axs[0, 0].plot(x, np.cos(x), 'g--', label='cos(x)')
axs[0, 0].legend()
axs[0, 0].set_title('Line Plot')

axs[0, 1].scatter(x, np.sin(x))
axs[0, 1].set_title('Scatter Plot')

axs[1, 0].hist(rand_data, bins = 30)
axs[1, 0].set_title('Histogram')

sc = axs[1, 1].scatter(rand_data, np.random.rand(1000), s = rand_data * 50, c = rand_data, cmap = 'viridis', alpha = 0.6)
fig.colorbar(sc, ax = axs[1, 1])
axs[1, 1].set_title('Colored Scatter')

plt.tight_layout()
plt.show()
```
