```python
import numpy as np
import time

def jacobi(A, b, x0=None, tol=1e-6, max_iterations=10000):
    n = len(b)
    if x0 is None:
        x = np.zeros_like(b)
    else:
        x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0=None, tol=1e-6, max_iterations=10000):
    n = len(b)
    if x0 is None:
        x = np.zeros_like(b)
    else:
        x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1
        x = x_new
    return x, max_iterations

def run_methods(A, b, tol=1e-6):
    x0 = np.zeros_like(b)
    
    # Jacobi
    start = time.time()
    x_jacobi, iter_jacobi = jacobi(A, b, x0, tol)
    time_jacobi = time.time() - start

    # Gauss-Seidel
    start = time.time()
    x_gs, iter_gs = gauss_seidel(A, b, x0, tol)
    time_gs = time.time() - start

    print("Jacobi solution:", x_jacobi)
    print("Jacobi iterations:", iter_jacobi)
    print("Jacobi time(s):", time_jacobi)
    print("-" * 60)
    print("Gauss-Seidel solution:", x_gs)
    print("Gauss-Seidel iterations:", iter_gs)
    print("Gauss-Seidel time(s):", time_gs)
    print("=" * 60)

print("System 1:")
A1 = np.array([
    [0.582745, 0.48 , 0.10 , 0.  , 0.  ],
    [0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
    [0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
    [0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
    [0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)
b1 = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
run_methods(A1, b1)

print("System 2:")
A2 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b2 = np.array([1, 1], dtype=float)
run_methods(A2, b2)

print("System 3:")
A3 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b3 = np.array([1, 1.0001], dtype=float)
run_methods(A3, b3)
```
