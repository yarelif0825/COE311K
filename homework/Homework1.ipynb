```python
#Homework 1 Monte Carlo Pi Approximation
import numpy as np
import matplotlib.pyplot as plt

#Monte Carlo Pi
def monte_carlo_pi(num_samples):
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    inside_circle = (x**2 + y**2) <= 1
    pi_estimate = 4 * np.sum(inside_circle)/num_samples
    return pi_estimate

#Iterate from 10 to 1000
sample_sizes = np.arange(10, 1001, 10)
pi_actual=np.pi
errors = []

for n in sample_sizes:
    pi_est = monte_carlo_pi(n)
    error = round(abs(pi_est - pi_actual), 7)
    errors.append(error)

#Plot Error
plt.figure(figsize=(8,5))
plt.plot(sample_sizes, errors, marker='o', linestyle='-')
plt.xlabel('Number of Samples')
plt.ylabel('Absolute Error (Est Pi - Actual Pi)')
plt.title('Error v Number of Samples')
plt.grid(True)
plt.show()
```
