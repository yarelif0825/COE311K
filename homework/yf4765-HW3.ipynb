```python
import numpy as np 
import matplotlib.pyplot as plt

#Taylor Series
def taylor_sin(x, n_terms):
    result = 0
    for n in range(n_terms):
        coeff = (-1)**n
        num = x**(2*n + 1)
        denom = np.math.factorial(2*n + 1)
        result += coeff * num / denom
    return result

# x values
x_vals = np.arange(0, 10.1, 0.1)
exact = np.sin(x_vals)

# Store errors
errors = []

for n_terms in range(1, 5):  # 1 to 4 terms
    approx = np.array([taylor_sin(x, n_terms) for x in x_vals])
    error = np.abs(approx - exact)
    errors.append(error)

# Plotting
plt.figure(figsize=(10, 6))
for i, error in enumerate(errors, 1):
    plt.plot(x_vals, error, label=f'{i} term(s)')
plt.title('Truncation Error in Taylor Series Approximation of sin(x)')
plt.xlabel('x')
plt.ylabel('Absolute Error')
plt.yscale('log')  # Log scale for error
plt.legend()
plt.grid(True)
plt.show()
```
