```python
import numpy as np
import matplotlib.pyplot as plt

# Function to estimate Pi using random sampling
def monte_carlo_pi(num_samples):
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    # Count points inside the unit circle
    inside = (x**2 + y**2) <= 1
    # Area ratio gives estimate for Pi
    return 4 * np.sum(inside) / num_samples

# Try different sample sizes from 10 to 1000, step 20
sample_sizes = np.arange(10, 1001, 20)
errors = []

for n in sample_sizes:
    pi_est = monte_carlo_pi(n)
    # Error to 15 digits past the decimal
    error = round(abs(pi_est - np.pi), 15)
    errors.append(error)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, errors, marker='o')
plt.xlabel('Number of Samples')
plt.ylabel('Absolute Error')
plt.title('Monte Carlo Pi Approximation Error')
plt.grid(True)
plt.show()
```
